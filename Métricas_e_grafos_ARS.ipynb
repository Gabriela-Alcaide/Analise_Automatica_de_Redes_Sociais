```python
import threading
import requests
```

```python
seguidores = []

url_seguidores = "https://bsky.social/xrpc/app.bsky.graph.getFollowers"

headers = {
    "Authorization": f"Bearer {token}"
}

lock = threading.Lock()

def get_seguidores_usuario(usuario):
  params = {
      "actor": usuario
  }
  retries = 5
  for attempt in range(retries):
    try:
      response = requests.get(url_seguidores, headers=headers, params=params)
      req.raise_for_status()
      with lock:
        seguidores.append({
            "usuario": usuario,
            "seguidores": response.json()
        })
      break
    except requests.exceptions.RequestException as e:
      if attempt < retries - 1:
        time.sleep(2 ** attempt)

threads = []
for h in base["handle"]:
  t = threading.Thread(target=get_seguidores_usuario, args=(h,))
  threads.append(t)
  t.start()

for t in threads:
  t.join()
```

```python
seguidores_tratados = []

for i in seguidores:
  nomes_seguidores = []
  if "followers" in i["seguidores"]:
    for s in i["seguidores"]["followers"]:
      nomes_seguidores.append(s["handle"])
  else:
    nomes_seguidores = []
  seguidores_tratados.append({
      "usuario": i["usuario"],
      "seguidores": nomes_seguidores
  })
```

```python
seguidores_tratados_rede = []

for i in seguidores_tratados:
  nomes_seguidores_rede = set(n for n in i["seguidores"] if n in base["handle"].values)
  seguidores_tratados_rede.append({
      "usuario": i["usuario"],
      "seguidores_rede": list(nomes_seguidores_rede)
  })
```

```python
import networkx as nx
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
```

```python
# Grafo: quem segue os usuários da sub-rede.
# A → B: A segue B.

G = nx.DiGraph()

for s in seguidores_tratados_rede:
  usuario = s["usuario"]
  seguidores = s["seguidores_rede"]

  # Adicionando o nó do usuário
  G.add_node(usuario)

  # Adicionando arestas para os seguidores
  for seguidor in seguidores:
    G.add_node(seguidor)  # Adiciona o nó do seguidor, se ainda não existir
    G.add_edge(seguidor, usuario)  # Cria uma aresta do usuário para o seguidor

node_color = []
highlight_nodes = []  # Lista para armazenar os nós que você deseja destacar

# Iterar sobre os nós do grafo
for n in G.nodes():
    linha = base[base["handle"] == str(n).strip()]
    if not linha.empty and 'classificacao' in linha.columns:
        if linha["classificacao"].iloc[0] == "bot":
            node_color.append("red")  # Cor padrão para bots
            highlight_nodes.append(n)  # Adiciona bots à lista de destaque
        else:
            node_color.append("steelblue")  # Cor padrão para humanos
    else:
        node_color.append("gray")  # Cor padrão se não houver classificação

# Desenhar o grafo
plt.figure(figsize=(10, 7))
pos = nx.spring_layout(G)  # Layout do grafo
nx.draw(G, pos, with_labels=False, node_size=100, node_color=node_color, font_size=10, font_weight='bold', arrows=True)

# Cria a legenda com cores para bots e humanos
bot_patch = mpatches.Patch(color='red', label='Bots')
human_patch = mpatches.Patch(color='steelblue', label='Humanos')

plt.legend(handles=[bot_patch, human_patch], loc="upper left", title="Legenda")

# Número de arestas.
num_arestas = G.number_of_edges()
print(f"Número de relacionamentos: {num_arestas}")

# Grau médio do grafo.
grau_medio = sum(dict(G.degree()).values()) / G.number_of_nodes() if G.number_of_nodes() > 0 else 0
print(f"Média de quantas contas da sub-rede um usuário segue: {grau_medio:.5f}")

# Densidade do grafo.
densidade = nx.density(G)
print(f"Densidade do grafo: {densidade:.5f}")

# Número de componentes fortemente conectados
num_componentes_fortes = nx.number_strongly_connected_components(G)
print(f"Número de componentes fortemente conectados: {num_componentes_fortes}")

# Número de componentes fracamente conectados
num_componentes_fracos = nx.number_weakly_connected_components(G)
print(f"Número de componentes fracamente conectados: {num_componentes_fracos}")

# Adiciona título e mostra o grafo
plt.title("Contas da sub-rede que seguem outras contas da sub-rede")
plt.savefig('grafo1.png', format="PNG")
plt.show()
```

```python
# Grafo: centralidade de grau.
# A → B: A segue B.

G = nx.DiGraph()

for s in seguidores_tratados_rede:
  usuario = s["usuario"]
  seguidores = s["seguidores_rede"]

  # Adicionando o nó do usuário
  G.add_node(usuario)

  # Adicionando arestas para os seguidores
  for seguidor in seguidores:
    G.add_node(seguidor)  # Adiciona o nó do seguidor, se ainda não existir
    G.add_edge(seguidor, usuario)  # Cria uma aresta do usuário para o seguidor

node_color = []
highlight_nodes = []  # Lista para armazenar os nós que você deseja destacar

# Iterar sobre os nós do grafo
for n in G.nodes():
    linha = base[base["handle"] == str(n).strip()]
    if not linha.empty and 'classificacao' in linha.columns:
        if linha["classificacao"].iloc[0] == "bot":
            node_color.append("red")  # Cor padrão para bots
            highlight_nodes.append(n)  # Adiciona bots à lista de destaque
        else:
            node_color.append("steelblue")  # Cor padrão para humanos
    else:
        node_color.append("gray")  # Cor padrão se não houver classificação

# Calcular as métricas
centralidade = nx.degree_centrality(G)

# Para dar destaque aos nós de interesse (por exemplo, aumentar o tamanho dos nós bots)
node_sizes = node_sizes = [10000 * centralidade[n] for n in G.nodes()]

# Desenhar o grafo
plt.figure(figsize=(10, 7))
pos = nx.spring_layout(G)  # Layout do grafo
nx.draw(G, pos, with_labels=False, node_size=node_sizes, node_color=node_color, font_size=10, font_weight='bold', arrows=True)

#for n in G.nodes():
 #   plt.text(pos[n][0], pos[n][1], f"{n}\nDeg In: {entrada_grau[n]}\nDeg Out: {saida_grau[n]}",
  #           fontsize=8, ha='center', va='center')

for n in G.nodes():
  if n == 'siteaudienciatv.bsky.social':
    plt.text(pos[n][0], pos[n][1] + 0.1, f"{n}\nCentralidade: {centralidade[n]:.2f}", fontsize=8, ha='center', va='center', color='blue', bbox=dict(facecolor='white', edgecolor='blue', boxstyle='round,pad=0.3', alpha=0.7))
  elif n == 'marcianobrito.bsky.social':
    plt.text(pos[n][0], pos[n][1] + 0.1, f"{n}\nCentralidade: {centralidade[n]:.2f}", fontsize=8, ha='center', va='center', color='blue', bbox=dict(facecolor='white', edgecolor='blue', boxstyle='round,pad=0.3', alpha=0.7))

# Cria a legenda com cores para bots e humanos
bot_patch = mpatches.Patch(color='red', label='Bots')
human_patch = mpatches.Patch(color='steelblue', label='Humanos')

plt.legend(handles=[bot_patch, human_patch], loc="upper left", title="Legenda")

# Adiciona título e mostra o grafo
plt.title("Centralidade de grau")
plt.savefig('grafo2.png', format="PNG")
plt.show()
```
